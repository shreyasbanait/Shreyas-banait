## Distribution of Dataset

```python
#Add at least one box and whisker plot of an important numerical variable in your table
import seaborn as sns
import matplotlib.pyplot as plt
important_variable = 'Sales per customer'

plt.figure(figsize=(8, 6))
plt.boxplot(data[important_variable], whis=1.5)
plt.xlabel(important_variable)
plt.title('Box Plot of Important Variable')
plt.show()
```

```python
#Add a correlation matrix of all your numerical variables and give written
#anlaysis of any variables that show strong correlation
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix = data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
strong_correlations = correlation_matrix[abs(correlation_matrix) >= 0.7].stack().reset_index()
strong_correlations = strong_correlations[strong_correlations['level_0'] != strong_correlations['level_1']]

print("Variables with strong correlation:")
print(strong_correlations)
```

```python
#BONUS: You can create a histogram to show distribution of a numerical variable

import seaborn as sns
import matplotlib.pyplot as plt

numerical_variable = 'Days for shipment (scheduled)'
plt.figure(figsize=(8, 6))
sns.histplot(data[numerical_variable], kde=True, color='blue')
plt.xlabel(numerical_variable)
plt.ylabel('Frequency')
plt.title(f'Distribution of {numerical_variable}')
plt.show()
```

## Aggregation of Categorical Data

```python
#Create either a bar chart or pie chart of the sum of one numerical value in
#relation to one categorical variable (i.e. sum of sales for each category of product)

import seaborn as sns
import matplotlib.pyplot as plt


numerical_value = 'Sales per customer'
categorical_variable = 'Category Name'


grouped_data = data.groupby(categorical_variable)[numerical_value].sum().reset_index()


plt.figure(figsize=(10, 6))
sns.barplot(y=categorical_variable, x=numerical_value, data=grouped_data, palette='Blues')


plt.xlabel(categorical_variable)
plt.ylabel('Sum of ' + numerical_value)
plt.title(f'Sum of {numerical_value} by {categorical_variable}')


```
